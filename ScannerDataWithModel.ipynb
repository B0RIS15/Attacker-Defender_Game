```python
'''
Есть три варианта совместить модель и данные от сканера
Первый вариант будет здесь
Мы получаем json ответ, в нем есть список уязвимостей(машин),
распределенных по "строгости". берем первое ненулевое
значение оттуда. это значение и есть наше количество 
уязвимостей(машин). Мы представляем что это набор уязвимостей
или машин(в оригинальной статье были машины, здесь будут уязвимости)
но моделе это абсолютно неважно. Далее проводим те же самые действия,
что и в оригинальной модели.
'''
import json
import random

class Machine:
    
    def __init__(self, vulnerable, percentValue, penetrate, protectionLevel, possibilityProtection):
        #Уязвима или нет - bool
        self.vulnerable = vulnerable
        #Значение в сети - %
        self.percentValue = percentValue
        #Взломана или нет -bool
        self.penetrate = penetrate
        #Защита - E' * percentValue
        self.protectionLevel = protectionLevel
        #Возможно ли ее защитить - bool
        self.possibilityProtection = possibilityProtection
        
    def changeVulnerable(self, vulnerable):
        self.vulnerable = vulnerable
    
    def changeCost(self, percentValue):
        self.percentValue = float(percentValue)
        
    def changePenetrate(self, penetrate):
        self.penetrate = penetrate
    
    def changeProtectionLevel(self, protectionLevel):
        self.protectionLevel = float(protectionLevel)
    
    def changePossibilityProtection(self, possibilityProtection):
        self.possibilityProtection = possibilityProtection
        
    def pr(self):
        print(self.vulnerable, self.percentValue, self.penetrate)

def JsonCheck():
    valueDefcon, valueCritical, valueHigh, valueMedium, valueLow, valueNegligible, valueUnknown = 0, 0, 0, 0, 0, 0, 0
    with open('new_clair_report.json') as f:
        file_content = f.read()
        templates = json.loads(file_content)
    f.close
    for i in templates["vulnerabilities"]:
        if i["severity"] == 'Defcon1':
            valueDefcon += 1
        elif i["severity"] == 'Critical':
            valueCritical += 1
        elif i["severity"] == 'High':
            valueHigh += 1
        elif i["severity"] == 'Medium':
            valueMedium += 1
        elif i["severity"] == 'Low':
            valueLow += 1
        elif i["severity"] == 'Negligible':
            valueNegligible += 1
        elif i["severity"] == 'Unknown':
            valueUnknown += 1
    print(valueDefcon, valueCritical, valueHigh, valueMedium, valueLow, valueNegligible, valueUnknown)
    commonValue = valueCritical + valueHigh + valueMedium + valueLow #пока не учитываем другие
    arrayValue = []
    arrayValue.append(valueCritical)
    arrayValue.append(valueHigh)
    arrayValue.append(valueMedium)
    arrayValue.append(valueLow)
    return arrayValue
    
def GetMachineValue(arrayValue):
    machineValue = 0
    for i in range(len(arrayValue)):
        if arrayValue[i] != 0:
            machineValue = arrayValue[i]
            break
    return machineValue

def GetVulnerableMachineValues(machineValues):
    vulnerableMachineValues = random.randint(1, machineValues)
    return vulnerableMachineValues

def GetImportantMachineValues(machineValues):
    importantMachineValues = random.randint(1, machineValues)
    return importantMachineValues

def GetVulnerableMachine(machineValues, machineArray):
    vulnerableMachine = 0
    for i in range(machineValues):
        if machineArray[i].vulnerable == True:
            vulnerableMachine += 1
    return vulnerableMachine

def GetPercentValue():
    percentValue = random.randint(10, 80)
    return percentValue

def GetMachineArray(machineValues, vulnerableMachineValues, importantMachineValues):
    machineArray = []
    percentValue = GetPercentValue()
    reversePercent = ((100 - percentValue) / (machineValues - 1)) / 100
    reversePercent = ("%.2f" % reversePercent)
    for i in range(machineValues):
        if i == (importantMachineValues - 1):
            tmp = Machine(False, float(percentValue / 100), False, float(0.0), False)
        else:
            tmp = Machine(False, float(reversePercent), False, float(0.0), False)
        machineArray.append(tmp)
    for i in range(0, vulnerableMachineValues):
        j = random.randint(0, machineValues - 1)
        machineArray[j].changeVulnerable(True)
    return machineArray
    
        
arrayValue = JsonCheck()
machineValues = GetMachineValue(arrayValue)
vulnerableMachineValues = GetVulnerableMachineValues(machineValues)
importantMachineValues = GetImportantMachineValues(machineValues)
machineArray = GetMachineArray(machineValues, vulnerableMachineValues, importantMachineValues)
for i in range(len(machineArray)):
    machineArray[i].pr()
print(GetVulnerableMachine(machineValues, machineArray))
```

```python
'''
Третий вариант - более приближен к оригинальной моделе
мы создаем несколько докер контейнеров, сканируем их -
какие то будут уязвимы, какие то нет, то есть сперва надо
просканировать неуязвимый контейнер, ну и еще надо организовать
единую папку для всех json'ов, чтобы модель получала данные из нее
'''
import glob
import os
import json
import random

class Machine:
    
    def __init__(self, vulnerable, percentValue, penetrate, protectionLevel, possibilityProtection):
        #Уязвима или нет - bool
        self.vulnerable = vulnerable
        #Значение в сети - %
        self.percentValue = percentValue
        #Взломана или нет -bool
        self.penetrate = penetrate
        #Защита - E' * percentValue
        self.protectionLevel = protectionLevel
        #Возможно ли ее защитить - bool
        self.possibilityProtection = possibilityProtection
        
    def changeVulnerable(self, vulnerable):
        self.vulnerable = vulnerable
    
    def changePercentValue(self, percentValue):
        self.percentValue = float(percentValue)
        
    def changePenetrate(self, penetrate):
        self.penetrate = penetrate
    
    def changeProtectionLevel(self, protectionLevel):
        self.protectionLevel = float(protectionLevel)
    
    def changePossibilityProtection(self, possibilityProtection):
        self.possibilityProtection = possibilityProtection
        
    def pr(self):
        print(self.vulnerable, self.percentValue, self.penetrate, self.protectionLevel, self.possibilityProtection)

def GetPercentValue():
    percentValue = random.randint(10, 80)
    return percentValue        

def GetMachineArray():
    try:
        os.chdir('report')
    except:
        os.chdir('/home/notebook/Attacker-Defender_Game/report')
    machineArray = []
    for file in glob.glob("*.json"):
        with open(file) as f:
            file_content = f.read()
            templates = json.loads(file_content)
        f.close
        try:
            templates["vulnerabilities"]
            tmp = Machine(True, float(0.0), False, float(0.0), False)
            machineArray.append(tmp)
        except:
            print('no vulnerabilities in report')
            tmp = Machine(False, float(0.0), False, float(0.0), False)
            machineArray.append(tmp)
    return(machineArray)

def AddingMachineArray(machineArray):
    lenArray = len(machineArray)
    importainMachineValue = random.randint(1, lenArray)
    percentValue = GetPercentValue()
    reversePercent = ((100 - percentValue) / (lenArray - 1)) / 100
    print(reversePercent)
    reversePercent = ("%.2f" % reversePercent)
    for i in range(lenArray):
        machineArray[i].changePercentValue(reversePercent)
    machineArray[importainMachineValue - 1].changePercentValue(percentValue / 100)
    
machineArray = GetMachineArray()
AddingMachineArray(machineArray)
for i in range(len(machineArray)):
    machineArray[i].pr()
```
